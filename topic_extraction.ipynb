```python
import meaningcloud
from time import sleep
import pandas as pd
import matplotlib.pyplot as plt
```

```python
meaningcloud_key = '6080c92cd8055f28dfc07b5619464baf'

def meaningcloud_analysis (text):  
    sleep (0.5)
    try:
        mc_topics =  meaningcloud.TopicsResponse(meaningcloud.TopicsRequest(meaningcloud_key, lang='en', txt=text).sendReq())
        mc_dict = mc_topics.getResults()
        results = mc_dict['concept_list']
        return results
        print (results)
    except:
        print("There was an error! "+ mc_topics.getStatusMsg())
        return None
    
```

```python
with open ("sound_project_1.txt", "r") as file:
    data=file.readlines()
```

```python
mc_output = meaningcloud_analysis(data)
```

```python
topics = []
for i in mc_output:
    topics.append(i['form'])
```

```python
freq = {} 
for items in topics: 
    freq[items] = topics.count(items) 
      
```

```python
freq
```

```python
topics_list=freq.keys()
count_list=freq.values()
```

```python
df_topics=pd.DataFrame()
```

```python
df_topics.head()
```

```python
df_topics.insert(loc=0, column = 'topic', value = topics_list)
```

```python
df_topics.insert(loc=0, column = 'freq', value = count_list)
```
